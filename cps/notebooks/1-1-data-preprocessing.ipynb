```python
# ════════════════════════════════════════════════
# TRY 2 — STANDARD SESSION OPENER
# ════════════════════════════════════════════════
import json, os, pickle, random, warnings
import numpy as np
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score
warnings.filterwarnings('ignore')

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

BASE = '/kaggle/input/datasets/aadityaupadhyay1353'

CFG_PATH = f'{BASE}/cps-config/config.json'

with open(CFG_PATH) as f:
    CFG = json.load(f)

TRAIN_IDS = CFG['engine_partition']['train']
VAL_IDS = CFG['engine_partition']['validation']
TEST_IDS = CFG['engine_partition']['test']

OUT = '/kaggle/working/'
os.makedirs(OUT, exist_ok=True)

print("TRY 2 session ready")
print("Train engines:", len(TRAIN_IDS))
print("Val engines:", len(VAL_IDS))
print("Test engines:", len(TEST_IDS))
print("Output path:", OUT)
```

```python
RAW_PATH = f'{BASE}/cmapss-turbofan/train_FD001.txt'

COLS = [
'engine_id','cycle','op1','op2','op3',
's1','s2','s3','s4','s5','s6','s7','s8','s9','s10',
's11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21'
]

df = pd.read_csv(RAW_PATH, sep=' ', header=None, index_col=False)
df.dropna(axis=1, how='all', inplace=True)

df.columns = COLS

print("Shape after load:", df.shape)
print("NaN count:", df.isnull().sum().sum())
print("Unique engines:", df['engine_id'].nunique())

DROP = CFG['sensors_to_drop']

df.drop(columns=DROP, inplace=True)

print("Active sensors:")
print([c for c in df.columns if c.startswith('s')])

mc = df.groupby('engine_id')['cycle'].max().reset_index()
mc.columns = ['engine_id','max_cycle']

df = df.merge(mc, on='engine_id')

df.to_csv(f'{OUT}v2_phase1_clean_dataset.csv', index=False)

print("Saved: v2_phase1_clean_dataset.csv")
print("Final shape:", df.shape)
```
